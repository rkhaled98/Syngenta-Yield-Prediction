```python
def create_grouping_df(end_point):
    df_test = df.loc[0:end_point]
#     df_test = df_test.drop_duplicates(subset=['PTT'])
    df_group = df_test.groupby('Yield')
    df_group_agg = pd.DataFrame(df_group.agg(lambda x: x.tolist()))
    df_group_agg = df_group_agg.drop(['PollenDap', 'PlantHeight', 'EarHeight'], axis=1)
    df_group_agg.Pedigree = df_group_agg.Pedigree.map(lambda x: x[0]) # get the first element in the Pedigree
    df_group_agg_ped_dummies = pd.get_dummies(df_group_agg.Pedigree)
    for idx, Series in df_group_agg.iterrows():
        if(np.asarray(Series.pID).shape[0] < 100):   
            df_group_agg = df_group_agg.drop(index=idx)
    df_group_agg = df_group_agg.drop(['Environment', 'Pedigree', 'pID'], axis=1) # previously had Pedigree dropped too
    return df_group_agg, df_group_agg_ped_dummies, df_group_agg.columns
```

```python
#THIS FUNCTION WILL TAKE INPUT DATAFRAME AND MAKE EACH ARRAY IN EACH (R0W,COLUMN) PART OF THE 
#FEATURES FOR THE TRAINING, USING THE FIRST NUM_DAYS_INC ENTRIES IN THE ARRAY.

def reframe_df(df, df_group_agg_ped_dummies, X_columns, NUM_DAYS_INC):
    END_NUMBER_OF_COLS = df.shape[1] * NUM_DAYS_INC
    df_new_cols = np.array([])

    # add NUM_DAYS_INC new columns for each X_columns
    for column in X_columns:
        for i in range(0,NUM_DAYS_INC):
            df_new_cols=np.append(df_new_cols, "%s%d" % (column, i))

    df_new = pd.DataFrame(columns=df_new_cols)
    row_yields = np.array([])


    for row_index, row in df.iterrows():
        row_yields = np.append(row_yields, df.loc[row_index].name)
        new_full_row = np.array([])
        for index, value in row.iteritems():
            value = np.asarray(value)
            value = value[0:NUM_DAYS_INC]
            new_full_row = np.append(new_full_row, value)
        new_full_row = new_full_row.reshape(1,END_NUMBER_OF_COLS)
        new_full_row = pd.DataFrame(new_full_row)
        new_full_row.columns = df_new_cols
        df_new = pd.concat([df_new, new_full_row], axis=0)
    df_new['End_Yield'] = row_yields
    df_new = df_new.reset_index(drop=True)
    df_group_agg_ped_dummies = df_group_agg_ped_dummies.reset_index(drop=True)
    df_new = pd.concat([df_new, df_group_agg_ped_dummies], axis=1)

    # df_new = pd.concat([df_new, row_yields], axis=1)
    return df_new

df_new = reframe_df(df_group_agg, df_group_agg_ped_dummies, X_columns, NUM_DAYS_INC=2)
```

```python
from sklearn.metrics import mean_squared_error
def RMSE(preds, y_test):
    return np.sqrt(mean_squared_error(preds, y_test))
```

```python
def get_feature_importances(regr, cols):
    feature_importances = regr.feature_importances_
    feature_importances = pd.Series(feature_importances)
    feature_importance_df = pd.DataFrame({'feature': cols,'feature_importance': feature_importances})
    feature_importance_df = feature_importance_df.sort_values(by=['feature_importance'])
#     for index, row in feature_importance_df.iterrows():
# #         print(row['feature'], 'has importance: ', row['feature_importance'])
#           print(row['feature'], ",", row['feature_importance'])
    return feature_importance_df
# get_feature_importances(regr)
```
